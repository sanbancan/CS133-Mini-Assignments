```python


## (1 point)
## Use ign_scores to determine which gaming platform is the best
## for racing game.

## Your code here . . .


import pandas as pd

# Load the IGN data
url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/ign_scores.csv"
ign_scores = pd.read_csv(url)

# Display the column names to inspect the structure
print(ign_scores.columns)

# Inspect the first few rows to see how genres are labeled
print(ign_scores.head())

# Filter the dataframe for the racing genre column
# Assuming the genre column for racing is named 'Racing', adjust if necessary
racing_scores = ign_scores[['Platform', 'Racing']]

# Drop rows with NaN values in the 'Racing' column
racing_scores = racing_scores.dropna(subset=['Racing'])

# Group by platform and calculate the average score for the 'Racing' genre
platform_avg_scores = racing_scores.groupby('Platform')['Racing'].mean()

# Identify the platform with the highest average score
best_platform = platform_avg_scores.idxmax()
highest_score = platform_avg_scores.max()

# Print the result
print(f"The best gaming platform for racing games is {best_platform} with an average score of {highest_score:.2f}.")


```

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from the URL
url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon_data = pd.read_csv(url)

# Filter the data for specific Type 1 values
filtered_data = pokemon_data[pokemon_data["Type 1"].isin(["Water", "Normal", "Grass", "Bug", "Psychic"])]

# Set the "Name" column as the index
filtered_data.set_index("Name", inplace=True)

# Select only numerical columns for the clustering
numerical_data = filtered_data[["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

# Create a color palette for the Type 1 annotations
type_colors = {
    "Water": "blue",
    "Normal": "gray",
    "Grass": "green",
    "Bug": "brown",
    "Psychic": "purple",
}
row_colors = filtered_data["Type 1"].map(type_colors)

# Generate the clustermap
sns.clustermap(
    numerical_data,
    row_cluster=True,
    col_cluster=False,
    figsize=(12, 8),
    cmap="coolwarm",
    row_colors=row_colors,
)

# Add a legend for the Type 1 colors
from matplotlib.patches import Patch
legend_patches = [Patch(color=color, label=type_name) for type_name, color in type_colors.items()]
plt.legend(handles=legend_patches, title="Type 1", loc="upper right", bbox_to_anchor=(1.2, 1))

plt.title("Clustermap of Pokémon by Selected Types", y=1.05)
plt.show()


```
