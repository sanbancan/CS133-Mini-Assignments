Here are the descriptions of this dataset:
- `#` - Pokedex entry number of the Pokemon
- `Name` - name of the Pokemon
- `Type 1` - each Pokemon has a type, this determines weakness/resistance to attacks [referred to as the primary type]
- `Type 2` - some Pokemon are dual type and have 2 [referred to as the secondary type]
- `Total` - sum of all stats that come after this, a general guide to how strong a Pokemon is
- `HP` - hit points, or health, defines how much damage a Pokemon can withstand before fainting
- `Attack` - the base modifier for normal attacks
- `Defense` - the base damage resistance against normal attacks
- `Sp. Atk` - special attack, the base modifier for special attacks
- `Sp. Def` - the base damage resistance against special attacks
- `Speed` - determines which Pokemon attacks first each round
- `Generation` - refers to which grouping/game series the Pokemon was released in
- `Legendary` - a boolean that identifies whether the Pokemon is legendary

## Q1: Plot the relationship between Attack and Defense
Create a relationship plot using seaborn's `replot` function. Plot `Attack` on the x-axis and `Defense` on the y-axis. Use `hue` to display the `HP` (hit point) for each datapoint. You can choose your favorite palette. Include a meaninful title in the plot. After creating the plot, insert a markdown cell immediately following the plot to describe the relationship between Attack and Defense in this dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon_data = pd.read_csv(url)

# Create the relationship plot
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=pokemon_data,
    x="Attack",
    y="Defense",
    hue="HP",
    palette="viridis",
    size="HP",
    sizes=(20, 200),
)

plt.title("Relationship Between Attack and Defense Stats of Pok√©mon", fontsize=16)
plt.xlabel("Attack", fontsize=19)
plt.ylabel("Defense", fontsize=19)
plt.legend(title="HP (Hit Points)", loc='upper right', bbox_to_anchor=(1.25, 1))
plt.grid(True)


plt.tight_layout()
plt.show()
```

From the plot, I observed....

Observations:

Positive Correlation: Pok√©mon with higher Attack values tend to also have higher Defense values, although there are some exceptions.

Diversity in HP: Pok√©mon with varying HP values are spread across the plot. Some Pok√©mon with high HP have either very high or low Attack and Defense.

Clustering: The data points show clusters around certain ranges, likely reflecting types or classifications of Pok√©mon (e.g., Legendary vs. non-Legendary).

Outliers: A few Pok√©mon have either extreme Attack or Defense values, standing out from the rest.

This visualization highlights how Pok√©mon with different characteristics (e.g., offensive vs. defensive specialization) are distributed across the combat stats.

## Q2: Plot the relationship between Attack and Total using a hexbin plot
Create a hexbin plot using seaborn's `joinplot` function. Plot `Attack` on the x-axis and `Total` on the y-axis. You can choose the gridsize. Include a title for the plot "Attack vs total for all Pokemon". After generating the plot, insert a markdown cell to answer this question: The hexbin plot displays the relationship between `Attack` and `Total`, what else does this plot show?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a hexbin plot using seaborn's jointplot
hexbin_plot = sns.jointplot(
    data=pokemon_data,
    x="Attack",
    y="Total",
    kind="hex",
    gridsize=25,
    cmap="coolwarm",
)

# Add a title
hexbin_plot.fig.suptitle("Attack vs Total for All Pok√©mon", fontsize=16)
hexbin_plot.fig.tight_layout()
hexbin_plot.fig.subplots_adjust(top=0.95)

# Show the plot
plt.show()
```

From the plot, I observed....

Observations:

The hexbin plot displays the following:

Relationship between Attack and Total:

There is a noticeable positive correlation; Pok√©mon with higher Attack values generally tend to have higher Total stats.
The higher the Attack, the greater the contribution to the overall strength (Total).

Density:

The plot highlights regions with high concentrations of Pok√©mon (indicated by warmer colors).
Most Pok√©mon have Attack values in the range of 50‚Äì100 and Total values around 400‚Äì600.
Outliers:

A few Pok√©mon stand out with exceptionally high Attack values (>150) or high Total values (>700).
The hexbin plot not only helps visualize the relationship between Attack and Total but also provides insights into the distribution and density of Pok√©mon stats, offering a clearer understanding of where most Pok√©mon fall in the dataset.








## Q3: Create two categorical plots for Pokemon types. `Type 1` and `Type 2` .
For each plot, the unique values of `Type 1` or `Type 2` should be plotted along the y-axis, and the counts of Pokemon should be on the x-axis. Use the color red for the `Type 1` plot and blue for the `Type 2` plot. After the two plots are displayed, insert a markdown cell to write your findings: Which are the most common `Type 1` and `Type 2` Pokemons, and which are the most rare ones?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(15, 10))

# Plot for Type 1
plt.subplot(1, 2, 1)
sns.countplot(
    data=pokemon_data,
    y="Type 1",
    color="red",
    order=pokemon_data["Type 1"].value_counts().index
)
plt.title("Count of Pok√©mon by Primary Type (Type 1)", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Primary Type", fontsize=12)

# Plot for Type 2
plt.subplot(1, 2, 2)
sns.countplot(
    data=pokemon_data,
    y="Type 2",
    color="blue",
    order=pokemon_data["Type 2"].value_counts().index
)
plt.title("Count of Pok√©mon by Secondary Type (Type 2)", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Secondary Type", fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()
```
